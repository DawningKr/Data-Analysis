```python
import matplotlib.pyplot as plt
from collections import defaultdict
from itertools import product
```

```python
def is_in_interval(number, minimum, maximum):
    return minimum <= number <= maximum

def generate_coin_sample(num_flips=10):
    weight_sample_space = defaultdict(int)
    for coin_flips in product(["Tails", "Heads"], repeat=num_flips):
        heads_count = len([outcome for outcome in coin_flips if outcome == 'Heads'])
        weight_sample_space[heads_count] += 1

    return weight_sample_space
```

```python
x = range(0, 10)
y = [2 * value for value in x]
plt.plot(x, y)
where = [is_in_interval(value, 2, 6) for value in x]
plt.scatter(x, y)
plt.fill_between(x, y, where=where)
plt.xlabel("Values between zero and ten")
plt.ylabel("Twice the value of x")
plt.show()
```

```python
weight_sample_space = generate_coin_sample()
weight_sample_space_20_flips = generate_coin_sample(20)
```

```python
sample_space_size = sum(weight_sample_space.values())
x_10_flips = list(weight_sample_space.keys())
y_10_flips = list(weight_sample_space[key] for key in x_10_flips)
prob_x_10_flips = [value / sample_space_size for value in y_10_flips]
where = [not is_in_interval(value, 3, 7) for value in x_10_flips]
plt.plot(x_10_flips, prob_x_10_flips)
plt.scatter(x_10_flips, prob_x_10_flips)
plt.fill_between(x_10_flips, prob_x_10_flips, where=where)
plt.xlabel("Head-count")
plt.ylabel("Probability")
plt.show()
```
